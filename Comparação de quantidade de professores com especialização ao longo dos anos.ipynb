## Índice de profissionais especializados da região sudeste


Demonstraremos, como base de estudo, a comparação do índice de profissionais especializados ao longo dos anos de 2015, 2016, 2017 e 2018. Demonstrando suas mudanças entre os anos e a quantidade apresentada de profissionais na região sudeste.

```python
import pandas as pd
import matplotlib as mtb
```

```python
df = pd.read_csv(r'/home/felippe/Documentos/DOCENTES_SUDESTE_2015.CSV', delimiter='|', encoding='iso-8859-1', usecols=['NU_ANO_CENSO', 'NU_IDADE', 'TP_ESCOLARIDADE', 'IN_POS_NENHUM'])
df16 = pd.read_csv(r'/home/felippe/Documentos/DOCENTES_SUDESTE_2016.CSV', delimiter='|', encoding='iso-8859-1', usecols=['NU_ANO_CENSO', 'NU_IDADE', 'TP_ESCOLARIDADE', 'IN_POS_NENHUM'])
df17 = pd.read_csv(r'/home/felippe/Documentos/DOCENTES_SUDESTE_2017.CSV', delimiter='|', encoding='iso-8859-1', usecols=['NU_ANO_CENSO', 'NU_IDADE', 'TP_ESCOLARIDADE', 'IN_POS_NENHUM'])
df18 = pd.read_csv(r'/home/felippe/Documentos/DOCENTES_SUDESTE_2018.CSV', delimiter='|', encoding='iso-8859-1', usecols=['NU_ANO_CENSO', 'NU_IDADE', 'TP_ESCOLARIDADE', 'IN_POS_NENHUM'])
```

```python
menor_30_ano_15 = df.query('TP_ESCOLARIDADE  == 4 & NU_IDADE < 30 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
maior_50_menor_70_ano_15 = df.query('TP_ESCOLARIDADE  == 4 & NU_IDADE <= 70 & NU_IDADE > 49 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
maior_30_menor_50_ano_15 = df.query('TP_ESCOLARIDADE  == 4 & NU_IDADE < 50 & NU_IDADE > 29 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
menor_30_ano_16 = df16.query('TP_ESCOLARIDADE  == 4 & NU_IDADE < 30 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
maior_30_menor_50_ano_16 = df16.query('TP_ESCOLARIDADE  == 4 & NU_IDADE < 50 & NU_IDADE > 29 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
maior_50_menor_70_ano_16 = df16.query('TP_ESCOLARIDADE  == 4 & NU_IDADE <= 70 & NU_IDADE > 49 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
menor_30_ano_17 = df17.query('TP_ESCOLARIDADE  == 4 & NU_IDADE < 30 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
menor_30_ano_18 = df18.query('TP_ESCOLARIDADE  == 4 & NU_IDADE < 30 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
maior_30_menor_50_ano_17 = df17.query('TP_ESCOLARIDADE  == 4 & NU_IDADE < 50 & NU_IDADE > 29 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
maior_30_menor_50_ano_18 = df18.query('TP_ESCOLARIDADE  == 4 & NU_IDADE < 50 & NU_IDADE > 29 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
maior_50_menor_70_ano_17 = df17.query('TP_ESCOLARIDADE  == 4 & NU_IDADE <= 70 & NU_IDADE > 49 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
maior_50_menor_70_ano_18 = df18.query('TP_ESCOLARIDADE  == 4 & NU_IDADE <= 70 & NU_IDADE > 49 & IN_POS_NENHUM == 0')[["NU_IDADE"]].count()
```

```python
obj = {
    'ano' : ['2015', '2016', '2017', '2018'],
    'Professores' : [int(menor_30_ano_15), int(menor_30_ano_16), int(menor_30_ano_17), int(menor_30_ano_18)]
}


df = pd.DataFrame(obj, index = ['2015','2016','2017','2018'])
p = df.plot(kind="bar")
p.set_xlabel("Faixa etária:abaixo de 30 anos ", fontsize=(16));
p.set_ylabel("Professores", fontsize=(16));
p.set_title("Número de professores com menos de 30 anos com pelo menos uma especialização", fontsize=(16))
```

No gráfico a cima, é possível observar o déficite de grande escala de professores com faixa etária a baixo de 30 anos. Com o passar dos anos, houve baixas com mais de 10 mil professores. A cada ano, menos profissionais estão buscando especializações. 

```python
obj = {
    'ano' : ['2015', '2016', '2017', '2018'],
    'Professores' : [int(maior_30_menor_50_ano_15), int(maior_30_menor_50_ano_16), int(maior_30_menor_50_ano_17), int(maior_30_menor_50_ano_18)]
}


df = pd.DataFrame(obj, index = ['2015','2016','2017','2018'])
p = df.plot(kind="bar", figsize=(8,8), yticks=(0,600000,800000,1000000,1100000))
p.set_xlabel("Faixa etária: 31 à 70 anos", fontsize=(16));
p.set_ylabel("Professores", fontsize=(16));
p.set_title("Número de professores com idade entre 31 à 50 anos com pelo menos uma especialização", fontsize=(16))
                
```

Já na faixa etária de 30 à 50 anos, é possível notar uma semelhança no número de professores especializados. Levando em consideração, que em todos os anos, mantiveram o número mínimo de 1 milhão de profissionais com pelo menos uma especialização, demonstrando que nessa faixa etária, está alocado o maior numéro de profissionais. 

```python
obj = {
    'ano' : ['2015', '2016', '2017', '2018'],
    'Professores' : [int(maior_50_menor_70_ano_15), int(maior_50_menor_70_ano_16), int(maior_50_menor_70_ano_17), int(maior_50_menor_70_ano_18)]
}


df = pd.DataFrame(obj, index = ['2015','2016','2017','2018'])
p = df.plot(kind="bar", figsize=(8,8))
p.set_xlabel("Faixa etária: 51 à 70 anos", fontsize=(16));
p.set_ylabel("Professores", fontsize=(16));
p.set_title("Número de professores com idade entre 51 à 70 anos com pelo menos uma especialização", fontsize=(16))
                
```

Os números de profissionais especializados neste gráfico, demonstra uma crescente na busca por especialização. Podemos observar um número menor do que a faixa etária de 30 à 40 anos, mas levando em consideração o grande número de profissionais que se aposentam próximo à essa faixa etária de 50 à 70 anos.

# Conclusão

#### Este trabalho nos demonstrou que prefessores com menos de 30 anos estão cada vez menos buscando uma especialização cedo, mas pudemos observar uma média padrão nos longos dos anos dos professores com faixa etária entre 31 à 50 anos. Essa queda de profissionais com especialização ao decorrer dos anos é preocupante, pois afeta diretamente no ensino das escolas.
